# Clustering
Below shows examples of clustering using hierarchical,CURE and DBSCAN clustering methods in kdb+/q.

### Dependencies
To run the below notebook, the following dependencies are required.
```
- embedPy
- matplotlib
- pyclustering
- scikit learn
```

### Example 1

````q
\l ml/ml.q
.ml.loadfile`:clust/init.q
\l graphics.q
plt:.p.import`matplotlib.pyplot
.p.import[`mpl_toolkits.mplot3d]`:Axes3D;
fcps:.p.import[`pyclustering.samples.definitions]`:FCPS_SAMPLES
read:.p.import[`pyclustering.utils]`:read_sample
````

````q
show df:key .ml.clust.i.dd
show lf:key .ml.clust.i.ld
````

````q
d1:(60#"F";",")0:`:data/ss5.csv
plt[`:scatter][;]. flip d1;
plt[`:show][];
````

````q
(cross/)(4;-1_df;5;0;01b)
````

````q
plot[`cure;d1;(cross/)(4;-1_df;5;0;01b)] /;0)]
````

````q
plot[`dbscan;d1;(cross/)(-1_df;5;5)]
````

````q
plot[`hc;d1;(cross/)(4;-1_df;-1_lf)]
````

````q
plot[`ward;d1;(4;`e2dist;`ward)]
````

````q
plot[`kmeans;d1;(cross/)(4;1000;1b;df)]
````

### Example 3

````q
dsc:{[n;r;s](r+n?s-r)*/:(cos;sin)@\:4*n?acos 0}
d3 :flip(-5 1)+(1 1.5)*dsc[1000;0;1.8],'dsc[2000;3.1;4.2],'dsc[4000;5.2;6.5]
d3,:flip(4 -1)+(1 8)*dsc[4000;0;1.]
d3@:neg[1000&count d3]?count d3
plt[`:scatter][;]. flip d3;
plt[`:show][];
````

````q
plot[`cure;d3;(cross/)(4;-1_df;40;0;01b)]
````

````q
plot[`dbscan;d3;(cross/)(-1_df;2;1.2)]
````

````q
plot[`hc;d3;(cross/)(4;-1_df;-1_lf)]
````

````q
plot[`ward;d3;(4;`e2dist;`ward)]
````

````q
plot[`kmeans;d2;(cross/)(5;20;1b;df)]
````
